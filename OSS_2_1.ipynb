```python
import pandas as pd
from pandas import Series, DataFrame

df = pd.read_csv('/content/sample_data/2019_kbo_for_kaggle_v2.csv')

print("======================= Project #2-1-1 Results =======================")
for year in range(2015, 2019):
  print(f"\n{year}년 : ")

  year_data = df[df['year'] == year]

  for category in ['H', 'avg', 'HR', 'OBP']:
    print(f"\nTop 10 Players in {category}:")

    year_data_sorted = year_data.sort_values(by=[category, 'batter_name'], ascending=[False, True])
    year_data_sorted['Rank'] = year_data_sorted[category].rank(ascending=False, method='min')

    top_10_players = year_data_sorted[year_data_sorted['Rank'] <= 10]
    print(top_10_players[['Rank', 'batter_name', category]])

print("\n")
print("======================= Project #2-1-2 Results =======================")
print("\n")

df_2018 = df[df['year'] == 2018]
max_war_position = df_2018[df_2018['cp'] != '지명타자'].groupby('cp')['war'].idxmax()

result_df = df_2018.loc[max_war_position, ['cp', 'batter_name', 'war']]
print(result_df)

print("\n")
print("======================= Project #2-1-3 Result =======================")
print("\n")

selected_columns = ['R', 'H', 'HR', 'RBI', 'SB', 'war', 'avg', 'OBP', 'SLG']
correlations = df[selected_columns].corrwith(df['salary'])
max_corr_column = correlations.abs().idxmax()
print(f"The element with the highest correlation with 'salary' is : {max_corr_column}")
```
